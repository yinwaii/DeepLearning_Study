```python
# -*- coding: utf-8 -*-
import numpy as np
%matplotlib inline
# 这是什么奇怪的操作，能优化性能？
# 还nm跑不通
from matplotlib import pyplot as plt
```

```python
np.random.seed(100)
x = np.linspace(-1, 1, 100).reshape(100, 1)
y = 3 * np.power(x, 2) + 2 + 0.2 * np.random.randn(x.size).reshape(100, 1)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
w1 = np.random.rand(1, 1)
b1 = np.random.rand(1, 1)
lr = 0.001
```

```python
for i in range(1000):
    y_pred = w1 * np.power(x, 2) + b1
    loss = np.sum(0.5 * np.power(y_pred - y, 2))
    grad_w = np.sum((y_pred - y) * np.power(x, 2))
    grad_b = np.sum(y_pred - y)
    w1 -= lr * grad_w
    b1 -= lr * grad_b
```

```python
plt.plot(x, y_pred, color = 'red', label = 'predict')
plt.scatter(x, y, color = 'blue', marker = 'o', label = 'true')
plt.xlim(-1, 1)
plt.ylim(2, 6)
plt.legend()
plt.show()
print(w1, b1)
```
